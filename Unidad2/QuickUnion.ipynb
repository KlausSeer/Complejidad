```python
from graphviz import Digraph, Source
```

```python
class Dottable:
    def Dot(self):
        dot = Digraph()
        for i in range(self.N):
            dot.node(str(i))
        for i in range(self.N):
            if self.sets[i] != i:
                dot.edge(str(i), str(self.sets[i]))
        dot.graph_attr['rankdir'] = 'BT'
        return dot
```

```python
class DisjointSetQU(Dottable):
    def __init__(self, n):
        self.N = n
        self.sets = [i for i in range(n)]
        
    def Find(self, a):
        x = a
        while x != self.sets[x]:
            x = self.sets[x]
        return x
    
    """def Find(self, x):
        if x == self.sets[x]:
            return x
        else:
            return self.Find(self.sets[x])"""
    
    def Union(self, a, b):
        pa = self.Find(a)
        pb = self.Find(b)
        if pa != pb:
            self.sets[pa] = pb
```
